```python
#import dataset ML model and data pre-processing tools

import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split  
from sklearn.datasets import load_boston 
from sklearn.preprocessing import StandardScaler
import xgboost as xgb

#Load Dataset

boston = load_boston()
house_pricing = pd.DataFrame(boston.data, columns=boston.feature_names)
print(house_pricing)
house_pricing["house_price"]=boston.target
print(house_pricing)
x,y = boston.data, boston.target

#split dataset

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
rows, coloums = x_test.shape
print( rows, coloums)   
rows, coloums = x_train.shape   
print(rows, coloums)  

#standardize

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

#train model

xgb_reg = xgb.XGBRegressor()
xgb_reg.fit(x_train_scaled, y_train)   
train_score = xgb_reg.score(x_train_scaled, y_train)
test_score = xgb_reg.score(x_test_scaled, y_test)

#check accuracy

print("Training R^2 score:", train_score)
print("Testing R^2 score:", test_score)

#feed new input
                             
new_input_values = []
for feature_name in boston.feature_names:
    value = float(input("Enter value for {}: ".format(feature_name)))
    new_input_values.append(value)


new_input = np.array(new_input_values).reshape(1, -1)  


new_input_scaled = scaler.transform(new_input)


prediction = xgb_reg.predict(new_input_scaled)


print("Predicted house price:", prediction[0])
```
